```python
import google.generativeai as genai
from google.colab import userdata
import PIL.Image
import os
```

# **Use Gemini API**

```python
from google.colab import userdata
userdata.get('GOOGLE_API_KEY')
if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

genai.configure(api_key=os.environ['GOOGLE_API_KEY'])
```

# **List all model**

```python
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(m.name)
```

# **To read image **

```python
image = PIL.Image.open('/content/Screenshot 2024-10-31 091642.png')
image
```

# **Use Gemini model to extract the detials from image**

```python
vision_model = genai.GenerativeModel('models/gemini-1.5-flash-8b-exp-0924')
```

# **Now Just Ask Question to Model to check the model**

```python
response = vision_model.generate_content(["What is total amount",image])
print(response.text)
```

# **Now convert image data into JSON format**

```python
response = vision_model.generate_content(["Convert the chart in image into data in json format?",image])
print(response.text)
```
